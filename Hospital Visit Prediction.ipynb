## Null Value Treatment

```python
# Numerical columns with null values
numcols=['age']
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='mean',verbose=1)
df[numcols]=si.fit_transform(df[numcols])
```

```python
X=df.drop('hospvis',axis=1)
y=df['hospvis']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
models=[]
tscore=[]
```

```python
def VIF(formula,data):
    from patsy import dmatrices
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    y , X = dmatrices(formula,data = data,return_type="dataframe")
    vif = pd.DataFrame()
    vif["VIF Factor"] = [variance_inflation_factor(X.values, i) \
       for i in range(X.shape[1])]
    vif["features"] = X.columns
    return(vif.round(1))

VIF=VIF("hospvis ~ docvis+ year+ edlevel+ age+\
       outwork+ female+ married+ kids+ hhninc+ educ+ self+\
       edlevel1+ edlevel2+ edlevel3+ edlevel4" ,data = df)
VIF
```

```python
vif_cols=VIF[VIF['VIF Factor']<10].features
```

# Feature selection

## RFE

```python
X=df.drop('hospvis',axis=1)
y=df['hospvis']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn.feature_selection import RFE

lr=LinearRegression()
rfe = RFE(lr,10)
rfe.fit(X_train,y_train)
print(rfe.support_)
print(rfe.ranking_)
idc_rfe = pd.DataFrame({"rfe_support" :rfe.support_,
                       "columns" :X_train.columns ,
                       "ranking" : rfe.ranking_,
                      })
rfecols = idc_rfe[idc_rfe["rfe_support"] == True]["columns"].tolist()
rfecols
```

## Forward Selection

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
sfs = SFS(lr, k_features='best', forward=True, floating=False, 
          scoring='neg_mean_squared_error', cv=10)
model = sfs.fit(X_train, y_train)

fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')

plt.title('Sequential Forward Selection (w. StdErr)')
plt.grid()
plt.show()
```

# Model Building

```python
X=df[rfecols]
y=df['hospvis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state=10)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression()
lr.fit(X_train,y_train)
# print(lr.score(X_train,y_train))
ypredtrain=lr.predict(X_train)
print('trainscore',r2_score(y_train,ypredtrain))
ypred = lr.predict(X_test)
print('testscore',r2_score(y_test,ypred))
models.append('LinearRegression_2')
tscore.append(r2_score(y_test,ypred))
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(
    criterion='mse',
    splitter='best',
    max_depth=5,
    min_samples_split=5,
    min_samples_leaf=5,
    min_weight_fraction_leaf=0.0,
    max_features=10,
    random_state=0,
    max_leaf_nodes=3,
    min_impurity_decrease=0.0)
dt.fit(X_train,y_train)
```

```python
print('trainscore',dt.score(X_train,y_train))
ypreddt = dt.predict(X_test)
print('testscore',r2_score(y_test,ypreddt))
models.append('DecisionTreeRegressor1')
tscore.append(r2_score(y_test,ypreddt))
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=None,
                      max_features='auto', max_leaf_nodes=5,
                      min_impurity_decrease=0.0, min_impurity_split=None,
                      min_samples_leaf=3, min_samples_split=4,
                      min_weight_fraction_leaf=0.0, n_estimators=13)

rf.fit(X_train,y_train)
```

```python
print('trainscore',rf.score(X_train,y_train))
ypredrf = rf.predict(X_test)
print('testscore',r2_score(y_test,ypredrf))
models.append('RandomForestRegressor1')
tscore.append(r2_score(y_test,ypredrf))
```

## Ridge

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

lambdas=np.linspace(1,100,100)

params={'alpha':lambdas}

model=Ridge(fit_intercept=True)

gridsearch=GridSearchCV(model,param_grid=params,cv=10,scoring='neg_mean_absolute_error')

gridsearch.fit(X_train,y_train)
```

```python
print('trainscore',abs(gridsearch.score(X_train,y_train)))
ypredgridsearch = gridsearch.predict(X_test)
print('testscore',r2_score(y_test,ypredgridsearch))
models.append('Ridge')
tscore.append(r2_score(y_test,ypredgridsearch))
```

## Lasso

```python
from sklearn.linear_model import Lasso
lambdas=np.linspace(1,10,100)
model=Lasso(fit_intercept=True)
params={'alpha':lambdas}
gridsearch2=GridSearchCV(model,param_grid=params,cv=10,scoring='neg_mean_absolute_error')
gridsearch2.fit(X_train,y_train)
```

```python
print('trainscore',abs(gridsearch2.score(X_train,y_train)))
ypredgridsearch2 = gridsearch2.predict(X_test)
print('testscore',r2_score(y_test,ypredgridsearch2))
models.append('Lasso')
tscore.append(r2_score(y_test,ypredgridsearch2))
```
